```python
import matplotlib.pyplot as plt

year = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
access = [84, 84, 83, 85, 85, 83, 86, 86, 88, 87]

plt.plot(year,access,marker="*")
plt.xlabel("Year")
plt.ylabel("Proportion of Families")
plt.title("Proportion of Families with Access to Improved Water")

plt.show()
```
