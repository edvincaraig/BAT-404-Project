```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.DataFrame({
    'Year': [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015],
    'Cases': [49000, 52000, 170000, 125000, 185000, 210000, 124000, 220000],
    'Death': [400, 500, 1080, 600, 900, 600, 430, 620]
})
ax = df["Cases"].plot(kind="bar", color="#09f7cf")
ax.set_ylabel("Number of Cases")
ax.set_xlabel("")
ax.set_ylim(0,250000)
for tick in ax.get_xticklabels():
  tick.set_rotation(0)

ax2 = ax.twinx()
ax2.plot(ax.get_xticks(),df["Death"], color = "#fc029c", linewidth = 4, marker = "o")
ax2.grid(False)
ax2.set_ylabel("Number of Deaths")
ax2.set_ylim(0,1200)

plt.title("Number of Cases and Death due to Dengue")
plt.show()
```
