```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Basic': [70, 78, 74],
    'Limited': [15, 16, 15],
    'Unimproved': [7, 3, 5],
    'Open Defacation': [8, 3, 6]
 })
dist = ["Rural","Urban","Total"]

ax = df.plot(stacked=True, kind='bar')
for bar in ax.patches:
    height = bar.get_height()
    width = bar.get_width()
    x = bar.get_x()
    y = bar.get_y()
    label_text = height
    label_x = x + width / 2
    label_y = y + height / 2
    ax.text(label_x, label_y, label_text, ha='center',    
            va='center')
    
ax.set_xticklabels(dist,rotation='horizontal')
plt.title("Percent Distribution of Families by Service Level of Sanitation Facilities")
plt.show()
```
