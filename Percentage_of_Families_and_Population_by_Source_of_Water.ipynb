```python
import matplotlib.pyplot as plt
import pandas as pd
  
data = {'Rural': [30,5,4,3.5,4,6,4.5,2.6,1.5,0.4,2],
        'Urban': [40.3,20.3,12,6.7,5.3,4.3,3.8,2.6,1.1,0.2,6]
       }
df = pd.DataFrame(data,columns=['Urban','Rural'], index = ['Water Station','Piped Dwelling','Tube Well','Spring','Piped to Yard','Public Tap','Piped to Neighbor','Bottled Water','Tanker Truck','Rainwater','Unimproved Source'])

plt.style.use('ggplot')

df.plot.barh()

plt.title('Percentage of Families and Population by Source of Water')
plt.ylabel('Rural')
plt.xlabel('Urban')
plt.show()
```
