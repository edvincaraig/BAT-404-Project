```python
import numpy as np
from matplotlib import pyplot as plt
print(plt.style.available)
plt.style.use('fivethirtyeight')

slices = [ 15, 6, 5, 74] #out of any number

labels = [ '15%', '6%', '5%', '74%']

colors = ['#00A36C', '#93C572', '#50C878', '#AFE1AF']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor' : 'white', 'linewidth': 1}, colors=colors)



plt.title("Percentage of Filipinos Practicing Sanitation Facilities")
plt.grid(True)
plt.show()
```
